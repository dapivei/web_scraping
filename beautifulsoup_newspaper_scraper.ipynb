```python
# Required libraries
import requests
from bs4 import BeautifulSoup
import numpy as np
import re
import pandas as pd
```

```python
#url = 'https://www.eldiario.net/noticias/2020/'
yesterday = '200501/'
today = '200502/'
url_yesterday = 'https://www.eldiario.net/noticias/2020/2020_05/nt'+ yesterday
url_today = 'https://www.eldiario.net/noticias/2020/2020_05/nt'+ today
```

```python
# Scraping 
r1 = requests.get(url_yesterday)
coverpage = r1.content
soup1 = BeautifulSoup(coverpage, 'html')
coverpage_news = soup1.find_all('h2')
number_of_articles = len(coverpage_news)
# Empty lists for content, links and titles
news_contents = []
list_links = []
list_titles = []
for n in np.arange(1, number_of_articles):

    # Getting the link of the article
    coverpage_news_links = soup1.find_all('a', class_='bloque')
    link = url_yesterday + coverpage_news_links[n] ['href'] 
    list_links.append(link)
    
    # Getting the title
    title = coverpage_news[n].get_text()
    list_titles.append(title)
    
    # Reading the content (it is divided in paragraphs)
    article = requests.get(link)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html')
    body = soup_article.find_all('p')
    
    # Unifying the paragraphs
    list_paragraphs = []
    for p in np.arange(0, len(body)):
        paragraph = body[p].get_text()
        list_paragraphs.append(paragraph)
        final_article = " ".join(list_paragraphs)
        
    final_article=re.sub("\\xa0", "", final_article)
    
    news_contents.append(final_article)
```

```python
df = pd.DataFrame({'Article Title': list_titles, 'Article Link': list_links, 'Article Content': news_contents})
```
